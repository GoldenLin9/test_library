```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
data = pd.read_excel('https://docs.google.com/spreadsheets/d/1s77beEKoEOVfvWVPfQ0GCzWa8Ctdzo-fot9Jzy3qkHQ/export')
data.head(12)
# read in the two sets of data
# if data need to be cropped, do that in a spreadsheet first
# give the dataframes different names when they get read in
```

```python
data.shape
```

```python
beta = pd.read_excel('https://docs.google.com/spreadsheets/d/1WeR6uQZyMTCDIte5XWnpn9kJAa4Hf3gFaEhuMxqkztg/export')
beta.head(33)
```

```python
data.describe()
```

```python
plt.hist(data['Accelleration'], bins = 10, label = "Free Fall", color = "purple")
plt.hist(beta['Accelleration'], bins = 10, label = "Accelerating Skateboard", color = "forestgreen")
plt.title("Free Fall vs Skateboard")
plt.xlabel("Acceleration")
plt.ylabel("Frequency")
plt.legend()
plt.show()
print("""
Initially, we used a sliding block and skateboard to compare the acceleration, however due to the shortage of data from the sliding block, we used a peice of foam.
We dropped the foam from a set height but because of our video quality with every video we took we found a seperate video online. 
This video contained a falling ball and measurements for how far it is. We used this video because a falling ball has a constant acceleration of 9.8. This made our graph more in line with the graph of the skateboard.
""")
```

```python
beta.shape
```

```python
beta.describe()
```

```python
nSkate = beta['Accelleration'].count()
mSkate = beta['Accelleration'].mean()
uncSkate = (beta['Accelleration'].std())/math.sqrt(nSkate)

nTrack = data['Accelleration'].count()
mTrack = data['Accelleration'].mean()
uncTrack = (data['Accelleration'].std()/math.sqrt(nTrack))

tHolmes = (mSkate-mTrack)/(math.sqrt(uncTrack**2 + uncSkate**2))
print(tHolmes)
```

Since our "t-value" is greater than 3, it shows that our data sets are most likely not similar. A recommendation for futurer experimentation would be picking videos of 2 balls with approximate acceleration so that it is easier to compare the data. Also, you could compare multiple objects like 3 or 4 balls to have more data
